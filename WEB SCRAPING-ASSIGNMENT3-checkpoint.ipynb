```python
#1. Write a python program which searches all the product under a particular product from www.amazon.in. The 
#product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
#guitars.



search_item = input("Enter the product to be searched: ")


driver = webdriver.Chrome()


driver.get("https://www.amazon.in/")


search_bar = driver.find_element(By.ID, "twotabsearchtextbox")
search_bar.send_keys(search_item)
search_bar.send_keys(Keys.RETURN)


time.sleep(5)


product_elements = driver.find_elements(By.CSS_SELECTOR, ".s-result-item")
for product_element in product_elements:
    try:
        product_name = product_element.find_element(By.CSS_SELECTOR, ".a-text-normal").text
        product_price = product_element.find_element(By.CSS_SELECTOR, ".a-price .a-offscreen").text
        print(f"Product: {product_name}")
        print(f"Price: {product_price}")
        print("-" * 50)
    except NoSuchElementException:
        
        continue


driver.close()
```

```python
#Q2In the above question, now scrape the following details of each product listed in first 3 pages of your search 
#results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
#scrape all the products available under that product name. Details to be scraped are: "Brand 
#Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
#“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time


def scrape_page(driver):
    product_details = []

   
    product_elements = driver.find_elements(By.CSS_SELECTOR, ".s-result-item")

    for product_element in product_elements:
        try:
            brand_name = product_element.find_element(By.CSS_SELECTOR, ".a-text-brand").text
        except:
            brand_name = "-"

        try:
            product_name = product_element.find_element(By.CSS_SELECTOR, ".s-title-instructions").text
        except:
            product_name = "-"

        try:
            price = product_element.find_element(By.CSS_SELECTOR, ".a-price .a-offscreen").text
        except:
            price = "-"

        try:
            return_exchange = product_element.find_element(By.CSS_SELECTOR, ".s-return-policy-ribbon").text
        except:
            return_exchange = "-"

        try:
            expected_delivery = product_element.find_element(By.CSS_SELECTOR, ".s-delivery-message").text
        except:
            expected_delivery = "-"

        try:
            availability = product_element.find_element(By.CSS_SELECTOR, ".s-availability").text
        except:
            availability = "-"

        try:
            product_url = product_element.find_element(By.CSS_SELECTOR, ".s-access-detail-page").get_attribute("href")
        except:
            product_url = "-"

        product_details.append([brand_name, product_name, price, return_exchange, expected_delivery, availability, product_url])

    return product_details


search_item = input("Enter the product to be searched on Amazon.in: ")


driver = webdriver.Chrome()

try:
    
    driver.get("https://www.amazon.in/")

    # Find the search bar and enter the search item
    search_bar = driver.find_element(By.ID, "twotabsearchtextbox")
    search_bar.send_keys(search_item)
    search_bar.send_keys(Keys.RETURN)

    
    all_product_details = []

 
    while True:
        time.sleep(5)  
        product_details = scrape_page(driver)
        all_product_details.extend(product_details)

        # Check if there is a next page
        next_page = driver.find_element(By.CSS_SELECTOR, ".s-pagination-next")
        if "s-pagination-disabled" in next_page.get_attribute("class"):
            break
        else:
            next_page.click()

 
    columns = ["Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Product URL"]
    df = pd.DataFrame(all_product_details, columns=columns)

  
    df.to_csv(f"{search_item}_products.csv", index=False)

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser
    driver.quit()
```

```python
#Q3 Write a python program to access the search bar and search button on images.google.com and scrape 10 
#images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’



import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By


def scrape_images(search_keyword, num_images):
    driver.get("https://www.google.com/imghp")

  
    search_bar = driver.find_element(By.NAME, "q")
    search_bar.send_keys(search_keyword)
    search_bar.send_keys(Keys.RETURN)

 
    for _ in range(num_images // 20):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

   
    image_elements = driver.find_elements(By.CSS_SELECTOR, ".rg_i")
    images = []

    for i, image_element in enumerate(image_elements[:num_images]):
        image_url = image_element.get_attribute("src")
        if image_url:
            images.append(image_url)

    return images


driver = webdriver.Chrome()  


keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
num_images_per_keyword = 10


all_images = {}
for keyword in keywords:
    images = scrape_images(keyword, num_images_per_keyword)
    all_images[keyword] = images


for keyword, images in all_images.items():
    print(f"Images for '{keyword}':")
    for i, image_url in enumerate(images, start=1):
        print(f"{i}. {image_url}")


driver.quit()
```

```python
#Q4 
#Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
#and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
#Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
#“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
#details is missing then replace it by “- “. Save your results in a dataframe and CSV. 



import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time


def scrape_smartphone_details(search_query):
    driver.get("https://www.flipkart.com/")


    search_bar = driver.find_element(By.NAME, "q")
    search_bar.send_keys(search_query)
    search_bar.send_keys(Keys.RETURN)


    time.sleep(5)


    product_elements = driver.find_elements(By.CSS_SELECTOR, "._1AtVbE")


    brand_names = []
    smartphone_names = []
    colors = []
    rams = []
    storage_roms = []
    primary_cameras = []
    secondary_cameras = []
    display_sizes = []
    battery_capacities = []
    prices = []
    product_urls = []

    for product_element in product_elements:
        try:
            brand_names.append(product_element.find_element(By.CSS_SELECTOR, "._4rR01T").text)
        except:
            brand_names.append("-")

        try:
            smartphone_names.append(product_element.find_element(By.CSS_SELECTOR, "._2WkVRV").text)
        except:
            smartphone_names.append("-")

        try:
            color = product_element.find_element(By.XPATH, ".//li[contains(text(), 'Color')]").text.split(":")[1].strip()
            colors.append(color)
        except:
            colors.append("-")

        try:
            ram = product_element.find_element(By.XPATH, ".//li[contains(text(), 'RAM')]").text.split(":")[1].strip()
            rams.append(ram)
        except:
            rams.append("-")

        try:
            storage_rom = product_element.find_element(By.XPATH, ".//li[contains(text(), 'ROM')]").text.split(":")[1].strip()
            storage_roms.append(storage_rom)
        except:
            storage_roms.append("-")

        try:
            primary_camera = product_element.find_element(By.XPATH, ".//li[contains(text(), 'Primary Camera')]").text.split(":")[1].strip()
            primary_cameras.append(primary_camera)
        except:
            primary_cameras.append("-")

        try:
            secondary_camera = product_element.find_element(By.XPATH, ".//li[contains(text(), 'Secondary Camera')]").text.split(":")[1].strip()
            secondary_cameras.append(secondary_camera)
        except:
            secondary_cameras.append("-")

        try:
            display_size = product_element.find_element(By.XPATH, ".//li[contains(text(), 'Display Size')]").text.split(":")[1].strip()
            display_sizes.append(display_size)
        except:
            display_sizes.append("-")

        try:
            battery_capacity = product_element.find_element(By.XPATH, ".//li[contains(text(), 'Battery Capacity')]").text.split(":")[1].strip()
            battery_capacities.append(battery_capacity)
        except:
            battery_capacities.append("-")

        try:
            price = product_element.find_element(By.CSS_SELECTOR, "._30jeq3").text
            prices.append(price)
        except:
            prices.append("-")

        try:
            product_url = product_element.find_element(By.CSS_SELECTOR, "._1fQZEK").get_attribute("href")
            product_urls.append(product_url)
        except:
            product_urls.append("-")

    smartphone_data = {
        "Brand Name": brand_names,
        "Smartphone Name": smartphone_names,
        "Colour": colors,
        "RAM": rams,
        "Storage(ROM)": storage_roms,
        "Primary Camera": primary_cameras,
        "Secondary Camera": secondary_cameras,
        "Display Size": display_sizes,
        "Battery Capacity": battery_capacities,
        "Price": prices,
        "Product URL": product_urls,
    }

    return smartphone_data


driver = webdriver.Chrome()  


search_query = input("Enter the smartphone you want to search for: ")

# Scrape smartphone details
smartphone_data = scrape_smartphone_details(search_query)


df = pd.DataFrame(smartphone_data)

# Save the DataFrame as a CSV file
df.to_csv(f"{search_query}_smartphones.csv", index=False)


driver.quit()
```

```python
#Q5  Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 


from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import re


def get_coordinates(city_name):
    
    driver = webdriver.Chrome()

    
    driver.get("https://www.google.com/maps")

    
    search_bar = driver.find_element(By.ID, "searchboxinput")
    search_bar.send_keys(city_name)
    search_bar.send_keys(Keys.RETURN)

    
    driver.implicitly_wait(10)

    
    map_url = driver.current_url

    
    driver.quit()

    
    match = re.search(r'@([-0-9.]+),([-0-9.]+)', map_url)
    if match:
        latitude = match.group(1)
        longitude = match.group(2)
        return latitude, longitude

    return None, None


city_name = input('Enter the city name: ')


latitude, longitude = get_coordinates(city_name)


if latitude and longitude:
    print(f'Latitude: {latitude}, Longitude: {longitude}')
else:
    print('Coordinates not found.')
```

```python
#Q6 Write a program to scrap all the available details of best gaming laptops from digit.in



import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
import time


driver_path = "path/to/chromedriver.exe"  
service = Service(driver_path)
driver = webdriver.Chrome()


url = 'https://www.digit.in/top-products/best-gaming-laptops-40.html'


driver.get(url)


brands = []
laptop_names = []
prices = []
ratings = []


laptop_elements = driver.find_elements(By.CLASS_NAME, 'TopNumbeHeading')


for laptop_element in laptop_elements:
    try:
        brand = laptop_element.find_element(By.CLASS_NAME, 'TopNumbeHeading').text.strip()
    except:
        brand = "-"

    try:
        laptop_name = laptop_element.find_element(By.CLASS_NAME, 'TopNumbeHeading').text.strip()
    except:
        laptop_name = "-"

    try:
        price = laptop_element.find_element(By.CLASS_NAME, 'top2020-price-price').text.strip()
    except:
        price = "-"

    try:
        rating = laptop_element.find_element(By.CLASS_NAME, 'rating-value').text.strip()
    except:
        rating = "-"

    brands.append(brand)
    laptop_names.append(laptop_name)
    prices.append(price)
    ratings.append(rating)


data = {'Brand': brands, 'Laptop Name': laptop_names, 'Price': prices, 'Rating': ratings}
df = pd.DataFrame(data)


df.to_csv('gaming_laptops.csv', index=False)


driver.quit()

print("Data has been scraped and saved as gaming_laptops.csv")
```

```python
#Q7 Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: 
#“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”



import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


driver_path = "path/to/chromedriver.exe"  
driver = webdriver.Chrome()


url = "https://www.forbes.com/billionaires/"


driver.get(url)

ranks = []
names = []
net_worths = []
ages = []
citizenships = []
sources = []
industries = []


scrolls = 5
for _ in range(scrolls):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)


billionaire_elements = driver.find_elements(By.CLASS_NAME, "data")


for billionaire_element in billionaire_elements:
    rank = billionaire_element.find_element(By.CLASS_NAME, "rank").text.strip()
    name = billionaire_element.find_element(By.CLASS_NAME, "personName").text.strip()
    net_worth = billionaire_element.find_element(By.CLASS_NAME, "netWorth").text.strip()
    age = billionaire_element.find_element(By.CLASS_NAME, "age").text.strip()
    citizenship = billionaire_element.find_element(By.CLASS_NAME, "countryOfCitizenship").text.strip()
    source = billionaire_element.find_element(By.CLASS_NAME, "source").text.strip()
    industry = billionaire_element.find_element(By.CLASS_NAME, "category").text.strip()

    ranks.append(rank)
    names.append(name)
    net_worths.append(net_worth)
    ages.append(age)
    citizenships.append(citizenship)
    sources.append(source)
    industries.append(industry)


data = {
    "Rank": ranks,
    "Name": names,
    "Net Worth": net_worths,
    "Age": ages,
    "Citizenship": citizenships,
    "Source": sources,
    "Industry": industries,
}
df = pd.DataFrame(data)


df.to_csv("forbes_billionaires.csv", index=False)

# Close the browser
driver.quit()

print("Data has been scraped and saved as forbes_billionaires.csv")
```

```python
#Q8 Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted 
#from any YouTube Video.

import time
from selenium import webdriver
from selenium.webdriver.common.by import By  


video_url = 'https://www.youtube.com/watch?v=ABwK03gZueg'


def scrape_comments():
  
    driver = webdriver.Chrome()  
    driver.get(video_url)
    time.sleep(5)  

  
    for _ in range(5):
        driver.execute_script('window.scrollTo(0, document.documentElement.scrollHeight);')
        time.sleep(3)

    comment_elements = driver.find_elements(By.CSS_SELECTOR, '#content-text')
    upvote_elements = driver.find_elements(By.CSS_SELECTOR, '#vote-count-middle')
    time_elements = driver.find_elements(By.CSS_SELECTOR, '.published-time-text')


    comments = [element.text for element in comment_elements]
    upvotes = [element.text for element in upvote_elements]
    timestamps = [element.text for element in time_elements]

   
    driver.quit()

    return comments, upvotes, timestamps


selenium_comments, selenium_upvotes, selenium_timestamps = scrape_comments()

for comment, upvote, timestamp in zip(selenium_comments, selenium_upvotes, selenium_timestamps):
    print(f"Comment: {comment}")
    print(f"Upvotes: {upvote}")
    print(f"Timestamp: {timestamp}")
    print("-" * 50)
```

```python
#Q9
#Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
#“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
#reviews, privates from price, dorms from price, facilities and property description. 


import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


driver = webdriver.Chrome()


url = 'https://www.hostelworld.com/pwa/wds/s?q=Goa,%20India&country=Goa&city=Goa&type=city&id=347&from=2023-10-03&to=2023-10-06&guests=2&page=1'
driver.get(url)

for _ in range(5):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.PAGE_DOWN)
    time.sleep(3)


hostel_elements = driver.find_elements(By.XPATH, "//div[@class='property-card']")

hostel_data_list = []


for hostel_element in hostel_elements:
    try:
        name = hostel_element.find_element(By.XPATH, ".//h2/a").text
    except:
        name = "-"

    try:
        distance = hostel_element.find_element(By.XPATH, ".//a[contains(@class, 'show-on-map')]//span[2]").text
    except:
        distance = "-"

    try:
        ratings = hostel_element.find_element(By.XPATH, ".//div[@class='rating']//span").text
    except:
        ratings = "-"

    try:
        total_reviews = hostel_element.find_element(By.XPATH, ".//a[@class='reviews']").text
    except:
        total_reviews = "-"

    try:
        overall_reviews = hostel_element.find_element(By.XPATH, ".//span[contains(@class, 'rating-rating')]").text
    except:
        overall_reviews = "-"

    try:
        privates_price = hostel_element.find_element(By.XPATH, ".//div[@class='price-col privates']//span").text
    except:
        privates_price = "-"

    try:
        dorms_price = hostel_element.find_element(By.XPATH, ".//div[@class='price-col dorms']//span").text
    except:
        dorms_price = "-"

    try:
        facilities = ", ".join([f.text for f in hostel_element.find_elements(By.XPATH, ".//span[contains(@class, 'facility-label')]")])
    except:
        facilities = "-"

    try:
        description = hostel_element.find_element(By.XPATH, ".//div[@class='rating-factors prop-card-tablet rating-factors small-text']//div[2]").text
    except:
        description = "-"

    hostel_data = {
        "Hostel Name": name,
        "Distance from City Centre": distance,
        "Ratings": ratings,
        "Total Reviews": total_reviews,
        "Overall Reviews": overall_reviews,
        "Privates from Price": privates_price,
        "Dorms from Price": dorms_price,
        "Facilities": facilities,
        "Property Description": description,
    }

    hostel_data_list.append(hostel_data)


if hostel_data_list:
    print(hostel_data_list[20])
else:
    print("No hostels found on the page.")




driver.quit()
```
